```python
#Part C
import pandas as pd
import numpy as np
import keras
```

```python
concrete_data = pd.read_csv('concrete_data.csv') #load data into the environment
concrete_data.head()
```

```python
concrete_data.isnull().sum() #check for null values
```

```python
concrete_data_columns = concrete_data.columns
# All columns other than 'Strength' variable are input variables
predictors = concrete_data[concrete_data_columns[concrete_data_columns != 'Strength']] # all columns except Strength
target = concrete_data['Strength'] # Strength column is traget variable
```

```python
predictors.head() #glimpse of the predictors
```

```python
target.head() #glimpse of the target variable
```

```python
#Normalization of predictors data
predictors_norm = (predictors - predictors.mean()) / predictors.std()
predictors_norm.head()
```

```python
n_cols = predictors_norm.shape[1] # number of predictors
print(n_cols)
```

```python
#Import nencesaary modules from keras
from keras.models import Sequential
from keras.layers import Dense
```

```python
#Building a regression model function definition
def regression_model():
  # create model
  model = Sequential()
  model.add(Dense(10, activation='relu', input_shape=(n_cols,))) #I1 Hidden layer
  model.add(Dense(1)) #Output layer

  # compile model
  model.compile(optimizer='adam', loss='mean_squared_error')
  return model
```

```python
#Step1: Test-Train split according to given constraints using sci-kit-learn library
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(predictors_norm, target, test_size=0.30, random_state=30)
```

```python
#Step2: Build the model
model = regression_model()
# Fit the model on train data
model.fit(X_train, y_train, epochs=100, verbose=2)
```

```python
#Predicting the output for the test dataset
predictions=model.predict(X_test)
print(len(predictions))
```

```python
#Step3: Calculate the MSE for the test datset
from sklearn.metrics import mean_squared_error
mse=mean_squared_error(y_test, predictions)
print(mse)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
mse_ls=[]
#Repeat steps(1-3) 50 times
for i in range (50):
  X_train, X_test, y_train, y_test = train_test_split(predictors_norm, target, test_size=0.30, random_state=30)
  model = regression_model()
  model.fit(X_train, y_train, epochs=100, verbose=2)
  predictions=model.predict(X_test)
  mse=mean_squared_error(y_test, predictions)
  mse_ls.append(mse)
```

```python
print(mse_ls) #list of 50 errors
```

```python
#Calculate the mean and standard deviation
import numpy as np
mean_errors=np.mean(mse_ls) #mean
std_errors=np.std(mse_ls) #standard deviation
print(mean_errors,std_errors) #----> Ans of Part C of assign
```
